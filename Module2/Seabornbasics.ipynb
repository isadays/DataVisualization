```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('clean_dd.csv')
```

```python
df.columns
```

```python
#Plot a histogram of snow depth
sns.histplot(data=df,x='SNWD')
```

```python
#Plot a scatter plot of snow fall with respect to temperature
sns.scatterplot(data=df, x='SNOW', y='TMAX')
```

```python
#Plot a line for each of the following: Minimum temperature, Observation temperature, Maximum temperature
sns.lineplot(data=df,x='DATE',y='TMAX')
```
